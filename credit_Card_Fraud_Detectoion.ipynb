```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
```

Thus this data is highly unbalanced

label 1=>fraudulant transaction

label 0=>normal transaction

```python
# we seperate the class 0 from the class 1
legit=d[d.Class==0]
fraud=d[d.Class==1]
```

#Building a sample data set with 492 legit values and 492 fraud values to make a good dataset

```python
legit_sample=legit.sample(n=492)
```

#concatinate the samples

```python
new_d=pd.concat([legit_sample,fraud],axis=0)
new_d.shape
```

```python
new_d.head()
```

```python
new_d.tail()
```

```python
new_d["Class"].value_counts()
```

```python
new_d.groupby("Class").mean()
#this step is v v imp 
#it helps us to figure out if out random selection is good since the values we get here should be close to the values we got on the entire dataset
```

```python
X=new_d.drop(columns="Class",axis=1)
Y=new_d["Class"]
```

```python
trainx,testx,trainy,testy=train_test_split(X,Y,random_state=2,test_size=0.2,stratify=Y)
```

```python
#Logistic Regression
m1=LogisticRegression()
```

```python
m1.fit(trainx,trainy)
```

#Accuracy Score using Logistic Regression

```python
predict_train1=m1.predict(trainx)
print("Accuracy Score of training data with Logistic Regression: ",accuracy_score(trainy,predict_train1))

predict_test1=m1.predict(testx)
print("Accuracy Score of testing data with Logistic Regression: ",accuracy_score(testy,predict_test1))
```

```python
#RandomForestClassifier

m2=RandomForestClassifier()
```

```python
m2.fit(trainx,trainy)
```

#Accuracy Score using RandomForestClassfier

```python
predict_train2=m2.predict(trainx)
print("Accuracy Score of training data with RandomForestClassifier: ",accuracy_score(trainy,predict_train2))

predict_test2=m2.predict(testx)
print("Accuracy Score of testing data with RandomForestClassifier: ",accuracy_score(testy,predict_test2))
```
